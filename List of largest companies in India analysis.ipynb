```python
df = pd.read_csv(r'C:\Users\sande\movies project\List of largest companies in India.csv', delimiter=',')

# Now, check the first few rows of the DataFrame to verify the data.
print(df.head())
```

```python
# Industry distribution
industry_counts = df['Industry'].value_counts()

# Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=industry_counts.index, y=industry_counts.values, palette='Set1')
plt.title('Industry Distribution of Top Companies in India')
plt.xlabel('Industry')
plt.ylabel('Number of Companies')
plt.xticks(rotation=90)
plt.show()

# Pie Chart
plt.figure(figsize=(14, 14))
industry_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('Set2', len(industry_counts)))
plt.title('Industry Distribution of Top Companies in India')
plt.ylabel('')
plt.show()
```

```python
# Scatter plot: Revenue vs Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Revenue(billions US$)', y='Profit(billions US$)', data=df, hue='Industry', palette='Set2', s=100)
plt.title('Revenue vs Profit Comparison')
plt.xlabel('Revenue (billions US$)')
plt.ylabel('Profit (billions US$)')
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Top 10 companies by market value
top_10_market_value = df.nlargest(10, 'Value(billions US$)')

plt.figure(figsize=(10, 6))
sns.barplot(x='Value(billions US$)', y='Name', data=top_10_market_value, hue='Name', palette='viridis', legend=False)
plt.title('Top 10 Companies by Market Value')
plt.xlabel('Market Value (billions US$)')
plt.ylabel('Company Name')
plt.show()
```

```python
# Bubble chart: Assets vs Market Value
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Assets(billions US$)', y='Value(billions US$)', data=df, size='Revenue(billions US$)', sizes=(20, 200), hue='Industry', palette='Set2', legend=None)
plt.title('Total Assets vs Market Value')
plt.xlabel('Total Assets (billions US$)')
plt.ylabel('Market Value (billions US$)')
plt.show()
```

```python
# Filtering the banks
banks = df[df['Industry'] == 'Banking']

# Sorting by Assets
top_banks_by_assets = banks.nlargest(10, 'Assets(billions US$)')

plt.figure(figsize=(10, 6))
sns.barplot(x='Assets(billions US$)', y='Name', data=top_banks_by_assets, hue='Name', palette='RdYlBu', legend=False)
plt.title('Top Banks by Assets')
plt.xlabel('Assets (billions US$)')
plt.ylabel('Bank Name')
plt.show()
```
