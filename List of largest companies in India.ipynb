```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_India#"

response = requests.get(url)
response
data = BeautifulSoup(response.content, "html.parser")
```

```python
table = data.find_all('table')[0]
table_titles = table.find_all('th')
table_titles
india_table_titles = [titles.text.strip() for titles in table_titles]
print(india_table_titles)
```

```python
df= pd.DataFrame(columns = india_table_titles)
```

```python
column_data = data.find_all('tr')
```

```python
for row in column_data[1:72]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)

    length = len(df)
    df.loc[length] = individual_row_data
```
